```python
df['cut'].unique()
```

```python
# creating a rank mapper according to the cut 
cut_map= {"Fair":1,"Good" :2,"Very Good":3,"Premium":4,"Ideal":5 }
```

```python
df['clarity'].unique()
```

```python
clarity_map = {'I1':1,'SI1':2,'SI2':3,'VS2':4,'VS1':5,'VVS2':6,'VVS1':7,"IF":8}
```

```python
df['color'].unique()
```

```python
color_map={'J':7,'I':6,'H':5,'G':4,'F':3,'E':2, 'D':1}
```

```python
# replacing the original values by ordinal ranks 
df['cut']= df['cut'].replace(cut_map)
```

```python
df['clarity']=df['clarity'].map(clarity_map)
```

```python
df['color']=df['color'].replace(color_map)
```
